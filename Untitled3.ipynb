```python
import copy

class Song:
           
    def __init__(self, name):
        
        self.song = midi.read_midifile(name)
        self.name = name
        self.res = self.song.resolution

            
    def extract_collection(self):
        self.song.make_ticks_abs()
        tracks = []
        states = list(np.zeros(128))
        for track in song:
            
            notes = [note for note in track if note.name == 'Note On' or note.name == 'Note Off']
            if len(notes) == 0:
                continue
                
            list_of_notes = []
            
            for note in notes:
                if note.name == 'Note On':
                    states[note.pitch] = note.tick
                else:
                    list_of_notes.append(Note(states[note.pitch], note.tick - states[note.pitch], note.pitch))
            
            tracks.append(sorted(list_of_notes, key = operator.attrgetter('tick')))
                

        return tracks
    

def cut_and_lengths(sg):
    
    res = sg.res
    GENERAL_BAG = []
    
    lgs = [4 * res, 2 * res, res, res / 2, res / 4, res / 8, res / 16]
    
    for track in sg.extract_collection():
        tr = copy.copy(track)
        
        i = 0
        while(len(tr) > 0):
            if (i == len(tr) - 1):
                if (len(tr) > 1):
                    GENERAL_BAG.append(tr)
                break
            if (tr[i].length >= lgs[0] + lgs[1]):
                if (i == 0):
                    tr = tr[1:]
                    i = 0
                else:
                    GENERAL_BAG.append(tr[:i])
                    tr = tr[i + 1:]
                    i = 0
            elif (tr[i].length >= lgs[1] + lgs[2]):
                tr[i].length = 1
                i = i + 1
                
            elif (tr[i].length >= lgs[2] + lgs[3]):
                tr[i].length = 2
                i = i + 1
                
            elif (tr[i].length >= lgs[3] + lgs[4]):
                tr[i].length = 4
                i = i + 1
            
            elif (tr[i].length >= lgs[4] + lgs[5]):
                tr[i].length = 8
                i = i + 1
                
            elif (tr[i].length >= lgs[5] + lgs[6]):
                tr[i].length = 16
                i = i + 1
                
            elif (tr[i].length >= lgs[6] * 1.5):
                tr[i].length = 32
                i = i + 1
            else:
                tr[i].length = 64
                i = i + 1
    
    return GENERAL_BAG
            
    
    
```

```python
class Note:
    def __init__(self,t,l,p):
        self.pitch = p
        self.length = l
        self.tick = t
```

```python
class Event:
    
    def __init__(self, ti, p, t):
        self.tick = ti
        self.pitch = p
        self.type = t
        
def create_track(seq):
    pattern = midi.Pattern()
    pattern.resolution = 240
    track = midi.Track()
    pattern.append(track)
    events = []
    
    for note in seq:
        events.append(Event(note.tick, note.pitch, "On"))
        events.append(Event(note.tick + 4 * pattern.resolution / note.length, note.pitch, "Off"))
    
    events = sorted(events, key = operator.attrgetter('tick'))
    
    for e in events:
        if e.type == "On":
            n = midi.NoteOnEvent(tick = e.tick, velocity = 20, pitch = e.pitch)
            track.append(n)
        else:
            n = midi.NoteOffEvent(tick = e.tick, velocity = 20, pitch = e.pitch)
            track.append(n)
    
    eot = midi.EndOfTrackEvent(tick=1)
    track.append(eot)
    
    print(track)

    midi.write_midifile("XXX.mid", pattern)
    
```

```python
def recover(diff, pitches, name):
    pattern = midi.Pattern()
    pattern.resolution = 15000
    track = midi.Track()
    pattern.append(track)
    
    t = 0
    for i in range(len(diff)):
        t = t + diff[i]
        on = midi.NoteOnEvent(tick = t, velocity = 20, pitch = pitches[i])
        track.append(on)
        
    eot = midi.EndOfTrackEvent(tick=1)
    track.append(eot)
    
    midi.write_midifile(name, pattern)
    
    return pattern

```
